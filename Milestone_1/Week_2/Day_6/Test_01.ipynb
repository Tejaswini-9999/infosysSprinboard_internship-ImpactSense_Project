## Practice on Pandas and Data Visualization

- 1. What is the overall survival rate of passengers aboard the Titanic?

```python
import pandas as pd
url='https://raw.githubusercontent.com/springboardmentor943x/ImpactSense-Intern-project/refs/heads/main/Milestone_1/Week_2/Day_6/titanic.csv'
df=pd.read_csv(url)
print(df['Survived'].mean()*100)
```

- 2. How does the survival rate vary by passenger class (Pclass)?

```python
survived_pclass=df.groupby('Pclass')['Survived'].mean()*100
print(survived_pclass)
```

- 3. What is the average age of passengers who survived vs. those who did not survive?

```python
age_passengers=df.groupby('Survived')['Age'].mean()
age_survived=age_passengers[1]
age_not_survived=age_passengers[0]
print(age_survived)
print(age_not_survived)
```

- 4. How many passengers embarked from each port (Embarked), and what are their survival rates?

```python
survival_port=df.groupby('Embarked')['Survived'].mean()*100
print(survival_port)
```

- 5. What is the distribution of fare prices paid by passengers, and does fare impact survival chances?

```python
import seaborn as sns
import matplotlib.pyplot as plt
#distribution
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df['Fare'], bins=50, kde=True)
plt.title("Distribution of Passenger Fares")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")

# impact
plt.subplot(1,2,2)  
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title("Fare vs Survival")
plt.xlabel("Survival (0 = Died, 1 = Survived)")
plt.ylabel("Fare")
plt.show()
```

- 6. Plot the count of passengers by passenger class (Pclass).

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Pclass', data=df)
plt.xlabel('Passenger Class')
plt.ylabel('Number of passengers')
plt.title('Passengers count in each class')
plt.show()
```

- 7. Visualize the survival rate with respect to gender using a bar chart.

```python
import numpy as np
sns.barplot(x='Sex',y='Survived',data=df,estimator=np.mean)
plt.xlabel("Sex")
plt.ylabel("Survival rate")
plt.title("Survival rate by Gender")
plt.show()
```

- 8. Show the age distribution of passengers using a histogram or KDE plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(data=df, x='Age',kde=True,color="orange")
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()
```

- 9. Create a boxplot comparing fares across different passenger classes.

```python
sns.boxplot(data=df,x='Pclass',y='Fare')
plt.xlabel('Passengr_class')
plt.ylabel('Fare')
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

- 10. Visualize survival count by embarkation port (Embarked) with a stacked bar chart.

```python
embarke_survived=pd.crosstab(df['Survived'],df['Embarked'])
embarke_survived.plot(kind='bar',stacked=True)
plt.xlabel('Survived')
plt.ylabel('Embarked')
plt.title('Embarked vs Survived')
plt.show()
```

- 11. Display a heatmap of the correlation matrix of numerical features in the dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corr_matrix = df.select_dtypes(include='number').corr()
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Heatmap of Titanic Numerical Features")
plt.show()
```
